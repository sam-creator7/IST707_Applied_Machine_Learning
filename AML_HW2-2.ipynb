```python
import pandas as pd

sales_df = pd.read_csv('supermarket_sales.csv') #reading csv file

sales_df.head() #printing first 5 rows to understand data
```

```python
#finding the sum of all missing values in each column - isnull() returns true or false and sum returns the total
sales_df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1.Gross income distribution over different branches
plt.figure(figsize=(10, 8))
sns.boxplot(data=sales_df, x='Branch', y='gross income')
plt.xlabel('Branch')
plt.ylabel('Gross Income')
plt.title('Gross Income Distribution Over Branches')
plt.show()
```

```python
# 2. Gender Differences in Each Branch
plt.figure(figsize=(8, 6))
sns.countplot(data=sales_df, x='Branch', hue='Gender')
plt.xlabel('Branch')
plt.ylabel('Count of Gender')
plt.title('Gender Differences in Each Branch')
plt.legend(title='Gender')
plt.show()
```

```python
#3. Customer Type Distribution in Each Branch
plt.figure(figsize=(8, 6))
sns.countplot(data=sales_df, x='Branch', hue='Customer type')
plt.xlabel('Branch')
plt.ylabel('Count of Cutomers')
plt.title('Customer Type Distribution in Each Branch')
plt.legend(title='Customer Type')
plt.show()
```

```python
#4. Product Line Distribution Across Branches
plt.figure(figsize=(10, 6))
sns.histplot(data=sales_df, x='Product line', hue='Branch', multiple='stack')
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Count')
plt.title('Product Line Distribution Across Branches')
plt.show()
```

```python
# separate DataFrames for each branch
branch_A = sales_df[sales_df['Branch'] == 'A']
branch_B = sales_df[sales_df['Branch'] == 'B']
branch_C = sales_df[sales_df['Branch'] == 'C']

# occurrences of each payment method using value_counts() - it returns unique value which will give us the total of each payment method
py_cnt_A = branch_A['Payment'].value_counts()
py_cnt_B = branch_B['Payment'].value_counts()
py_cnt_C = branch_C['Payment'].value_counts()

print("A",py_cnt_A)
print("B",py_cnt_B)
print("C",py_cnt_C)

# subplots for each branch
plt.figure(figsize=(15, 5))

# Branch A
plt.subplot(131)
plt.pie(py_cnt_A, labels=py_cnt_A.index, autopct='%1.1f%%', startangle=90)
plt.title('Payment Method Branch A')

# Branch B
plt.subplot(132)
plt.pie(py_cnt_B, labels=py_cnt_B.index, autopct='%1.1f%%', startangle=90)
plt.title('Payment Method Branch B')

# Branch C
plt.subplot(133)
plt.pie(py_cnt_C, labels=py_cnt_C.index, autopct='%1.1f%%', startangle=90)
plt.title('Payment Method Branch C')

plt.tight_layout()
plt.show()
```
